**La ecuación de Antoine (presión de vapor)**

Calcula la presión de vapor de una sustancia

$$long_{10}(P)=A-\frac{B}{C+T}$$

Constantes de Antoine para el agua (T en °C, P en mmHg)

A = 8.07131

B= 1730.63

```python
A = 8.07131

B= 1730.63

C= 233.426

T = 80 # °C

P = 10 ** (A-(B/(C+T)))
print(f"Presión de vapor del agua a {T} °C es {P: .2f} mmHg")

```

***EJEMPLO 3***

El número de Reynolds Re permite saber si el flujo es laminar, de transición o turbulento.

$$Re=\frac{pvD}{m}$$

rho = 1000 densidad [kg/m³]

v = 1.2 velocidad [m/s]

D = 0.05 diámetro [m]

mu = 0.001 viscosidad [Pa·s]

```python
rho = 1000   # densidad [kg/m³]
v = 1.2      # velocidad [m/s]
D = 0.05     # diámetro [m]
mu = 0.001   # viscosidad [Pa·s]

Re = (rho * v * D) / mu
print(f"Número de Reynolds = {Re:.0f}")
```

**3. Conversión de un reactor**


Verificación de conversión de una reacción. En un reactor, la conversión X indica qué tan completa está la reacción:

$$ X =\frac{FA0-FA}{FA0}$$

```python
FA0 = 30.0   # mol/s (flujo inicial de A)
FA = 13.0     # mol/s (flujo final de A)

X = (FA0 - FA) / FA0 * 100
print(f"Conversión: {X:.1f}%")

if X < 60:
    print("Conversión baja: ajustar condiciones del reactor.")
elif 60 <= X < 90:
    print("Conversión moderada: proceso aceptable.")
else:
    print("Conversión alta: operación eficiente.")
```

***EJERCICIO:*** **Monitoreo de temperatura en un reactor**

En un reactor continuo se mide la temperatura de operación cada cierto tiempo.
Por seguridad, el reactor no debe operar por encima de 80 °C.

El operador registra temperaturas hasta que el proceso se detiene.

Usar bucles, condicionales y listas para:

registrar datos experimentales,
analizarlos,
y tomar decisiones de operación.
Escribe un programa en Python que haga lo siguiente:

Pida al usuario temperaturas del reactor (en °C) una por una.
Guarde cada temperatura en una lista.
El programa debe seguir pidiendo valores mientras la temperatura sea menor o igual a 80 °C.
Si el usuario introduce una temperatura mayor a 80 °C, el proceso se detiene y:
se muestra un mensaje de alarma.

```python
# Lista donde se almacenarán las temperaturas válidas
temperaturas = []

# Inicializamos la variable temperatura
temperatura = 0

# Bucle while: se ejecuta mientras la temperatura sea segura
while temperatura <= 80:

    # Se pide la temperatura al usuario
    temperatura = float(input("Ingrese la temperatura del reactor (°C): "))

    # Condición de seguridad del proceso
    if temperatura <= 80:
        temperaturas.append(temperatura)  # Se guarda la temperatura
    else:
        print("ALARMA: Temperatura fuera del rango seguro.")
        print("Proceso detenido.")
print("\nTemperaturas registradas:", temperaturas)
```

**TUPLAS**

Una tupla es una estructura de datos inmutable y ordenada que permite almacenar una colección de elementos. Los elementos de una tupla se encierran entre paréntesis (), separados por comas.

Para crear una tupla, encierra los elementos entre paréntesis:

```python
vector = (3,4,1)
```

NO SE MIDIFICA EL VALOR POSICIÓN
Para acceder a los elementos de una tupla, utiliza el índice del elemento entre corchetes, similar a las listas:

```python
print(vector[1])  # Imprime la posición 1 de la tupla
```

A diferencia de las listas, las tuplas son inmutables, lo que significa que no se pueden modificar una vez creadas. No se pueden agregar, eliminar o cambiar elementos en una tupla existente.

Las tuplas son útiles cuando necesitas **almacenar una colección de elementos que no deben modificarse**, como coordenadas o datos de configuración.

**MÉTODOS DE TUPLAS**

Aunque las tuplas son inmutables, Python proporciona varios métodos útiles para trabajar con ellas:

***count(elemento):*** devuelve el número de veces que aparece un elemento en la tupla.

***index(elemento):*** devuelve el índice de la primera aparición de un elemento en la tupla. Opcionalmente, se puede especificar el inicio y fin de la búsqueda.

***len(tupla):*** aunque no es un método de tupla propiamente dicho, esta función incorporada devuelve la longitud de la tupla.

**DOCUMENTACIÓN DE FUNCIONES (docstrings)**

```python
def area_rectangulo(base, altura):
    """Calcula el área de un rectángulo.
    Args:
        base (float): La base del rectángulo.
        altura (float): La altura del rectángulo.
    Returns:
        float: El área del rectángulo.
    """
    return base * altura
```

```python
import re
```

```python
def limpiar_cadenas(strings):
    result = []
    for value in strings:
        value = value.strip()  # suprime blancos u otro carácter especificado del final
                               #o del principio de una expresión
        value = re.sub('[!*#?]', '', value) #reemplaza todas las ocurrencias de un patrón
                                           #por el texto que especifiquemos
        value = value.title() #convierte la primera letra de cada palabra de una cadena
                              # a mayúsculas.
        result.append(value)
    return result
```

**Ejercicio validación de contraseña segura para calentar motores**

Una contraseña se considera segura si cumple con los siguientes criterios:

* Tiene al menos 8 caracteres.
* Contiene al menos una letra mayúscula.
* Contiene al menos una letra minúscula.
* Contiene al menos un número.
* Contiene al menos un símbolo.

Escribe un código que pida al usuario ingresar una contraseña segura, si la contraseña ingresada no cumple con alguno de los criterios, el programa deberá indicarlo y solicitar nuevamente la contraseña.


```python
while True: #para repetir hasta que la contraseña sea válida.
    password = input("Ingresa una contraseña segura: ")

    tiene_mayuscula = False #contadores para verificar cada condición
    tiene_minuscula = False
    tiene_numero = False
    tiene_simbolo = False

    for caracter in password:
        if caracter.isupper(): #letra mayúscula
            tiene_mayuscula = True
        elif caracter.islower(): #letra minúscula
            tiene_minuscula = True
        elif caracter.isdigit(): #número
            tiene_numero = True
        else:
            tiene_simbolo = True

    if (len(password) >= 8 and
        tiene_mayuscula and
        tiene_minuscula and
        tiene_numero and
        tiene_simbolo):
        print("Contraseña segura")
        break # se sale del bucle solo si la condición es válida
    else:
        print("La contraseña no cumple con los criterios de seguridad.")
        print("Debe tener al menos 8 caracteres, una mayúscula, una minúscula, un número y un símbolo.\n")

```

#**ANÁLISIS ESTADÍSTICO**

**¿QUÉ ES EL ANÁLISIS ESTADÍSTICO DE DATOS?**

El análisis estadístico de datos es el conjunto de técnicas que usamos para describir, resumir e interpretar datos, con el fin de entender qué está pasando, comparar grupos y, en algunos casos, hacer inferencias o predicciones.

**1. ¿PARA QUÉ SIRVE?**
* ***Resumir*** grandes cantidades de datos en información clara.
* ***Detectar patrones*** (tendencias, diferencias entre grupos, relaciones).
* ***Tomar decisiones*** basadas en evidencia (por ejemplo, en ingeniería, economía, salud, educación).
* ***Evaluar hipótesis:*** comprobar si un cambio o una diferencia es “real” o puede explicarse por azar.


**2. TIPOS DE ANÁLISIS ESTADÍSTICOS**

***A) Estadística descriptiva***

Responde preguntas como:

* ¿Cuál es el valor “típico”?
* ¿Qué tan dispersos están los datos?
* ¿Hay valores muy altos o muy bajos (outliers)?

Herramientas comunes:

* ***Medidas de tendencia central:*** media, mediana, moda
* ***Medidas de dispersión:*** rango, varianza, desviación estándar, IQR
* ***Tablas y gráficos:*** histogramas, boxplots, barras, pastel (según el caso)

***B) Estadística inferencial (saca conclusiones sobre una población)***

Se usa cuando tienes una muestra y quieres concluir algo sobre una población. Incluye:

* Intervalos de confianza
* Pruebas de hipótesis (p-values)
* Comparaciones entre grupos (t-test, ANOVA)
* Relaciones entre variables (correlación, regresión)

#**3. PASOS TÍPICOS DE UN ANÁLISIS ESTADÍSTICO**

* ***Plantear el objetivo***

Ejemplo: “Comparar si hay diferencia entre regiones”

* ***Revisar y limpiar datos***

valores nulos, errores, duplicados, formatos

* ***Explorar y describir***

tablas resumen y estadísticas básicas

* ***Visualizar***

gráficos para entender distribución y comparaciones

* ***Interpretar resultados***

qué significa en el contexto del problema

* ***(Opcional) Inferir o modelar***

pruebas estadísticas o modelos predictivos

**4. CONCEPTOS CLAVE**

* ***Variable:*** característica medida (ej. Total, Hombres, Mujeres)
* ***Población vs muestra:*** todos los datos posibles vs una parte
* ***Distribución:*** forma en la que se reparten los valores
* ***Outliers:*** valores atípicos que pueden afectar conclusiones cuando hay estos datos se utiliza la mediana

**5. EJEMPLOS DE PREGUNTAS QUE SE RESPONDEN**

* ¿Cuál es el promedio del total por estado?
* ¿Qué estados tienen los valores más altos?
* ¿Hay diferencias en el total entre regiones?
* ¿Existe relación entre Hombres y Mujeres en los datos?


***Idea central:***

El análisis estadístico convierte datos en información útil para describir y tomar decisiones con fundamento.

#**PAQUETERÍA  STATISTICS EN PYTHON**

La paquetería ***statistics*** es un módulo incluido en la librería estándar de Python (no necesitas instalar nada) que sirve para realizar estadística descriptiva básica de manera rápida y clara.

Permite:

***1) Resumir datos experimentales***

* Media (mean)
* Mediana (median)
* Moda (mode) y multimoda (multimode)

***2) Medir variabilidad (dispersión)***

Te ayuda a cuantificar qué tanto “se dispersan” tus mediciones:

* Varianza muestral (variance)
* Desviación estándar muestral (stdev)
* Versiones poblacionales (pvariance, pstdev)

Esto es muy útil para evaluar precisión, ruido experimental y estabilidad del proceso.

***3) Calcular cuantiles y apoyar la detección de outliers***

Con quantiles() puedes obtener cuartiles (Q1, Q2, Q3) para construir:

* IQR (rango intercuartílico)
* Criterio para detectar valores atípicos (outliers)

**Nota**

* Cuando trabajas con listas pequeñas o medianas.
* Para análisis más avanzado (grandes datasets, DataFrames, gráficos, matrices), suele complementarse con NumPy y Pandas.




***EJERCICIO**

**MEDIDA DE TENDENCIA CENTRAL**

* Media (promedio)
* Mediana
* Moda


```python
import statistics as st
```

```python
datos= [7.2, 7.8, 6.8, 6.2, 8.2, 8, 8.2, 5.6, 8.6, 7.1, 8.2, 7.7, 7.5, 7.2, 7.7, 5.8, 6.8, 6.8, 8.5, 7.5, 6.1,7.9, 9.4, 9, 7.8, 8.5, 9, 7.7, 6.7, 7.7 ]
```

```python
print("Media:", st.mean(datos))
print("Mediana:", st.median(datos))

# Moda: si no hay una moda clara, puede lanzar StatisticsError

print("Moda:", st.mode(datos)) #Si no hay un valor que más se repite puede a dar un error
```

**MEDIDAS DE DISPERSIÓN**

* Varianza
* DE
* Rango

```python
# Varianza y desviación estándar (muestrales)
print("Varianza (muestral):", st.variance(datos))
print("Std (muestral):", st.stdev(datos))

# poblacionales (si fuera TODA la población)
print("Varianza (poblacional):", st.pvariance(datos))
print("Std (poblacional):", st.pstdev(datos))


print("Rango:", max(datos) - min(datos)) #valor maximo de mis datos-valor minimo, entre mayor rango tengo mas dispersión de mis datos y entre mas pequeño mis datos son mas centrales
```

**CUARTILES E IQR**


image.png

```python
# n=4 devuelve tres cortes: Q1, Q2, Q3
q1, q2, q3 = st.quantiles(datos, n=4, method="inclusive")
iqr = q3 - q1

print("Q1, Q2, Q3:", q1, q2, q3)
print("IQR:", iqr)
```

**DETECCIÓN DE OUTLIERS CON IQR**

```python
li = q1 - 1.5 * iqr
ls = q3 + 1.5 * iqr
outliers = [x for x in datos if x < li or x > ls]

print("Límite inferior:", li)
print("Límite superior:", ls)
print("Outliers:", outliers)
```

**COEFICINETE DE VARIACIÓN (CV)**

Indica qué tan dispersos están los datos en relación con su promedio.
Se interpreta como un porcentaje: “cuánta variación típica hay respecto a la media”.


$$CV=\frac sx  x  100%$$


x = media

s = desviación estándar(generalmente muestral)

**Interpretación**

* ***CV pequeño*** → datos más consistentes (menor variación relativa).
* ***CV grande*** → datos más dispersos (mayor variación relativa).

```python
cv = st.stdev(datos) / st.mean(datos) * 100
print("CV (%):", cv)
```

***Nota:*** CV solo tiene sentido cuando la media es positiva y no está cerca de cero, porque si la media es muy pequeña el CV puede salir enorme y engañoso.

#**USO DE NumPy PARA ESTADÍSTICA DE DATOS (ESTADÍSTICA DESCRIPTIVA)**

NumPy es fundamental para trabajar con arreglos numéricos (ndarray).
En estadística descriptiva, permite calcular de forma rápida y eficiente medidas como media, mediana, varianza, desviación estándar, percentiles y detección de outliers.


***1) ¿Por qué usar NumPy en estadística?***

* Es más rápido que trabajar con listas (usa operaciones vectorizadas).
* Facilita cálculos estadísticos con funciones ya implementadas.
* Permite trabajar con datos en forma de arrays, ideales para análisis científico.

***2) Convertir una lista a ndarray***


```python
import numpy as np

```

```python
datos = [7.2, 7.8, 6.8, 6.2, 8.2]
a = np.array(datos)
```

```python
a = np.array(datos)

n = a.size # tamaño del conjunto de datos
media = a.mean()
mediana = np.median(a) ###¿?

# Varianza y desviación estándar
var_pob = a.var(ddof=0)      # poblacional
std_pob = a.std(ddof=0)

var_mues = a.var(ddof=1)     # muestral
std_mues = a.std(ddof=1)

minimo = a.min()
maximo = a.max()
rango = maximo - minimo


print("media: ", media)
print("mediana: ", mediana)
print("varianza:", var_pob)
print("DE: ", std_pob)
print("rango: ", rango)
```

**CUARTILES Y PERCENTILES**

```python
q1, q2, q3 = np.percentile(a, [25, 50, 75]) #  Cuartiles, percentiles e IQR
iqr = q3 - q1

li = q1 - 1.5 * iqr # Outliers por criterio IQR
ls = q3 + 1.5 * iqr
outliers = a[(a < li) | (a > ls)]


print("\nCuartiles:")
print(f"Q1 = {q1:.3f}, Q2 = {q2:.3f}, Q3 = {q3:.3f}")
print(f"IQR = {iqr:.3f}")
print("Outliers =", outliers)
```

**COEFICINETE DE VARIACIÓN (CV)**

```python
cv = (std_mues / media) * 100
print(f"\nCoeficiente de variación (CV) = {cv:.2f}%")
```
